## Pivot() method
[souce](https://youtu.be/WcDaZ67TVRo?t=12025)

```python
#read dataset
df_gdp = pd.read_csv('./Assets/gdp.csv',delimiter=',',encoding='latin-1')
df_gdp.head(4)
```

```python
#Pivot, 
# index => column init rows
# columns , items header
df_gdp.pivot(index='year',columns='country',values='gdppc').head(5).round(0)
```

## **Pivot_table()** Method
[souce](https://youtu.be/WcDaZ67TVRo?t=12055)

### Example N°**1**

```python
#read dataset
dfSales = pd.read_excel('./Assets/supermarket_sales.xlsx')
dfSales.head(3)
```

```python
#Make pivot TB and add aggregate function
dfSales.pivot_table(
    index='Gender', #pivot by witch column
    aggfunc='sum', #agregation
    values=['Quantity','Total'], #columns to show
    columns='Product line' #columns heads
    ).round(0)
```

### Example N°**2**
Pivot table also can use for transform data and turn it on proper format. [video source](https://youtu.be/WcDaZ67TVRo?t=12798)

```python
dfPop = pd.read_csv('./Assets/population_total.csv')
dfPop
```

```python
#droping null values
dfPop.dropna(inplace=True #save changes in data
)
```

```python
#making pivot table
dfPop_pivot = dfPop.pivot(index='year',columns='country',values='population')
```
